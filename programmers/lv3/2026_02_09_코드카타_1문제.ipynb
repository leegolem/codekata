```python
def solution(number):
    answer = 0
    n = len(number)
    
    for i in range(n-2):
        for j in range(i+1, n-1):
            for k in range(j+1, n):
                if number[i] + number[j] + number[k] == 0:
                    answer+=1
    return answer
```

```python
def solution(t, p):
    answer = 0
    len_p = len(p)
    int_p = int(p)
    for i in range(len(t) - len_p + 1):
        if int(t[i:i+len_p]) <= int_p:
            answer += 1
    
    return answer
```
